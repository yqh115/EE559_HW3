Extract features from original datasets

```python

import numpy as np
import pandas as pd

b = np.zeros((1, 42))
file_name = 'dataset1.csv'
a1 = pd.read_csv(file_name, header=None, usecols=[0])
a2 = pd.read_csv(file_name, header=None, usecols=[1])
a3 = pd.read_csv(file_name, header=None, usecols=[2])
a4 = pd.read_csv(file_name, header=None, usecols=[3])
a5 = pd.read_csv(file_name, header=None, usecols=[4])
a6 = pd.read_csv(file_name, header=None, usecols=[5])

b[0][0] = a1.min()
b[0][1] = a1.max()
b[0][2] = a1.mean()
b[0][3] = a1.median()
b[0][4] = a1.std()
b[0][5] = a1.quantile(0.25)
b[0][6] = a1.quantile(0.75)

b[0][7] = a2.min()
b[0][8] = a2.max()
b[0][9] = a2.mean()
b[0][10] = a2.median()
b[0][11] = a2.std()
b[0][12] = a2.quantile(0.25)
b[0][13] = a2.quantile(0.75)

b[0][14] = a3.min()
b[0][15] = a3.max()
b[0][16] = a3.mean()
b[0][17] = a3.median()
b[0][18] = a3.std()
b[0][19] = a3.quantile(0.25)
b[0][20] = a3.quantile(0.75)

b[0][21] = a4.min()
b[0][22] = a4.max()
b[0][23] = a4.mean()
b[0][24] = a4.median()
b[0][25] = a4.std()
b[0][26] = a4.quantile(0.25)
b[0][27] = a4.quantile(0.75)

b[0][28] = a5.min()
b[0][29] = a5.max()
b[0][30] = a5.mean()
b[0][31] = a5.median()
b[0][32] = a5.std()
b[0][33] = a5.quantile(0.25)
b[0][34] = a5.quantile(0.75)

b[0][35] = a6.min()
b[0][36] = a6.max()
b[0][37] = a6.mean()
b[0][38] = a6.median()
b[0][39] = a6.std()
b[0][40] = a6.quantile(0.25)
b[0][41] = a6.quantile(0.75)

print(b)

writer = pd.ExcelWriter('newdata.xlsx', index=False)
df1 = pd.DataFrame(data=b)
df1.to_excel(writer, 'Sheet1')
writer.save()
writer.close()
```
